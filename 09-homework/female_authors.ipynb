```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text) 
```

```python
title = []
books = doc.find_all('a', class_='bookTitle')
for book in books:
    title.append(book.text.strip())
```

```python
author = []
writers = doc.find_all('a', class_='authorName')
for writer in writers:
    author.append(writer.text)
```

```python
ranking = []
ranks = doc.find_all('td', class_='number')
for rank in ranks:
    ranking.append(rank.text)
```

```python
doc.find('span', class_='smallText uitext').find('a').get_text(strip=True)
```

```python
score = []
vote = []
score_votes = doc.find_all('span', class_='smallText uitext')
for score_vote in score_votes:
    scores = score_vote.find_all('a')[0].text.strip()
    votes = score_vote.find_all('a')[1].text.strip()
    score.append(scores)
    vote.append(votes)
```

```python
rate = []
ratings = doc.find_all('span', class_='greyText smallText uitext')
for rating in ratings:
    rate.append(rating.text.strip())
    
    
    
    
    
    
    
    
    
```

```python
df = pd.DataFrame(list(zip(title, author, ranking, score, vote, rate)),columns=['title','author', 'ranking', 'score', 'vote','rating']) 
```
