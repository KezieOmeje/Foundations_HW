````python
import requests
from bs4 import BeautifulSoup
````

````python
response = requests.get("http://www.mineral.k12.nv.us/pages/School_Board_Minutes")
doc = BeautifulSoup(response.text)
````

````python
date = []
minutes = doc.find_all('p')
for minute in minutes[20:80]:
    date.append(minute.text)
    
````

````python
pdf_link = []
urls = doc.find_all('a')
for url in urls[22:65]:
    links = url['href']
    pdf_link.append(links)
````

````python
import pandas as pd
````

````python
df = pd.DataFrame(list(zip(minutes, pdf_link)),columns=['minutes','pdf_link']) 
df
````
