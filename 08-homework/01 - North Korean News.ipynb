```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("http://kcna.kp/kcna.user.home.retrieveHomeInfoList.kcmsf")
doc = BeautifulSoup(response.text)
```

```python
title = []
articles = doc.find_all('h3')
for article in articles:
    title.append(article.text)
        
```

```python
onclick = []
links = doc.find_all('a')
for link in links[27:146]:
    click = link.attrs
    onclick.append(click['onclick'])
    

```

```python
article_id = []
links = doc.find_all('a')
for link in links[27:146]:
    click = link.attrs
    article_id.append(click['onclick'][16:25])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(title, onclick, article_id)),columns=['article_title','onclick', 'article_id']) 
```

```python
df.to_csv('nk-news.csv', index=False)
```
